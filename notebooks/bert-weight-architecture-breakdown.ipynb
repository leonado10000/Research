```python
from safetensors.torch import load_file

# Load safetensors file
tensors = load_file(r"C:\Users\rjdis\OneDrive\Documents\GitHub\Research\artifacts\exported\bert-pretrained\model.safetensors")

# Basic checks
print(tensors.keys())
print(tensors["bert.embeddings.word_embeddings.weight"].shape)


def print_pretty_state_dict(state_dict):
    from collections import defaultdict

    # Tree builder
    def tree():
        return defaultdict(tree)

    root = tree()

    # Build nested structure
    for name, tensor in state_dict.items():
        parts = name.split(".")
        node = root
        for p in parts[:-1]:
            node = node[p]
        node[parts[-1]] = list(tensor.shape)

    # Pretty printing
    def print_tree(node, indent="", last=True, key=None):
        branch = "â””â”€â”€ " if last else "â”œâ”€â”€ "
        if key is not None:
            if isinstance(node, dict):
                print(f"{indent}{branch}{key}")
            else:
                print(f"{indent}{branch}{key:<30} {node}")

        indent += "    " if last else "â”‚   "

        if isinstance(node, dict):
            keys = list(node.keys())
            for i, k in enumerate(keys):
                print_tree(node[k], indent, i == len(keys) - 1, k)

    print("\nðŸ“¦ MODEL WEIGHTS (from safetensors)\n")
    print_tree(root)


# Pretty-print the safetensors weights
print_pretty_state_dict(tensors)
```
