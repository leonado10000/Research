```python
from transformers import Trainer, TrainingArguments
import torch
import torch.nn as nn
from torch.utils.data import Dataset

# ---- 1. Model ----
class F(nn.Module):
    def __init__(self):
        super().__init__()
        self.f = nn.Linear(5, 1)
        self.loss_fn = nn.MSELoss()

    def forward(self, input_ids=None, labels=None):
        preds = self.f(input_ids)
        loss = self.loss_fn(preds, labels)
        return {"loss": loss, "logits": preds}


model = F()

# ---- 2. Dataset ----
class SmallDS(Dataset):
    def __init__(self):
        self.x = torch.tensor([[1,2,3,4,5]] * 5, dtype=torch.float)
        self.y = torch.tensor([[i] for i in range(5)], dtype=torch.float)

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        return {
            "input_ids": self.x[idx],
            "labels": self.y[idx]
        }

ds = SmallDS()

# ---- 3. Trainer ----
args = TrainingArguments(
    output_dir="./out",
    per_device_train_batch_size=2,
    num_train_epochs=5,
    logging_steps=1,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=ds
)

trainer.train()
```
